```
# 스트링
dan= input('단을 입력하세요')
print(type(dan))  # 입력값은 스트링으로 저장됨

 # eval() 함수로 감싸면 숫자로 처리됨
dan= eval(input('단을 입력하세요'))
print(type(dan))
```

```
# 구구단 출력
dan= input('단을 입력하세요')
print(type(dan))  # 입력값은 스트링으로 저장됨
d= int(dan) # int(스트링) -> 스트링을 숫자로 변환
print(type(d))

for i in range(1,10):
    # print(dan,"x",i,"=",dan*i)
    print(d,"x",i,"=",d*i)  # int형
```

```
# print() 함수
# 문자열 연결
# 따옴표로 감싼 문자열은 + 연산과 동일함

print('good','morning')
# 문자열의 띄어쓰기는 컴머를 사용

# 10줄로 출력됨
# range(10) : 0~9

for i in range(10):
    print(i)  # 개행문자가 포함
# 한줄로 출력됨
for i in range(10):
    print(i,end=' ') # 개행문자 대신 공백문자 사용
```

```
# args : 명령행 매개변수

import sys
args= sys.argv[1:]
for i in args:
    print(i.upper(),end=' ')
```

```
name= input('이름을 입력하세요')
job= input('직업을 입력하세요')
print('Hello,{0}'.format(name, job))
```

```
import math
print(math.pi)
print('원주율:{0: f}'.format(math.pi))  # 소수점 6째 자리수까지 출력
# 소수 이하 2자리
print('원주율:{0: .2f}'.format(math.pi))
```

```
# 진수법
num= 10
str= '10진수:{0:d}, 16진수(소문자): {0:x}, 16진수(대문자): {0:X}'.format(num)
print(str)
str= '10진수:{0:d}, 16진수(소문자): {0:#x}, 16진수(대문자): {0:#X}'.format(num)
print(str)
```

```
# 천단위 컴머(,)
num= 123456789
print('{0:,}'.format(num))

# 백분률
num= 0.9815
print('{0:.0%}'.format(num))
print('{0:.2%}'.format(num))
```
