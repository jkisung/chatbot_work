```
# if문
color= input('색상을 입력하세요')
if color =='blue':
    # pass -> 들여쓰기를 안할 경우
    print('건너가세요')
else:
    print('기다리세요')
```

```
num= 99
# 숫자를 2로 나눈 나머지가 0이면
if num%2== 0:
    print("짝수입니다>")
else:
    print("홀수입니다.")
```

```
money= 1000
if money:
    print('택시를')
    print('타고')
    print('간다')
```

```
color= input('색상을 입력하세요')

if color == 'blue':
    print('건너가세요')
elif color == 'red':
    print('잠시만 기다리세요')
else:
    print('잘못된 색상입니다.')
```
