## List (리스트)
  - [] 대괄호 사용
  - range() 함수를 통한 인덱스 리스트 생성 가능

## Tupple (튜플)
  - 리스트와 유사하지만 튜플 내 값을 변경 못함
  - 각 값에 대해 인덱스가 부여됨
  - 변경 불가능
  - () 소괄호 사용

## dictionary (딕셔너리)
  - 정수형 인덱스가 아닌 키를 이용하여 값을 저장 및 조회
  - 저장된 각 자료에 대한 순서가 없음
  - { } 중괄호를 사용
  - , (컴머)를 기준으로 아이템 구분
  - 사전의 원소 -> key : value (쌍으로 이루어짐)
  - 사전.key() -> key만  추출 (임의 순서)
  - 사전.value() -> value만 추출 (임의 순서)
  - 사전.items() -> key와 value를 튜플로 추출 (임의 순서)
  - keys, values, items가 변환하는 자료형 --> 리스트
   

```python
# 리스트
# [] 대괄호로 감싸고 쉼표로 구분
a= [1,3,5,7,9]
b= a # 참조변수만 복사됨 (실제 리스트 복사는 아님)
b[0]= 10
print(a)
print(b)
```

```python
a= [1,2,3,4,5]
b= a[:]  # 실제 리스트가 복사됨
print(a)
print(b)
```

```python
a= [1,2,3,4,5]
print(len(a)) # 리스트의 요소 개수
# 슬라이스 [start:end-1]
print(a[0:2])  # 0번에서 1번까지

a= [1,2,3]
b= a*3  # 리스트 3회 반복
print(b)
```

```python
a= [1,2,3]
# print(a[2]+'hi')  # 숫자와 문자열은 + 할 수 없음
print(str(a[2])+'hi')

print(a[1:2])  # 리스트 1번 추출
print(type(a[1:2]))

print(a[1]) 
print(type(a[1]))

a[1:2]= ['a','b','c']  # 리스트의 내용 변경
print(a)

a= [1,2,3]
a[1]= ['a','b','c']  # 하위 리스트로 입력됨
print(a)
```

```python
a= [1,2,3]
a[1:2] = ['a','b','c']  # 리스트의 내용 변경
print(a)
```

```python
# 리스트의 삭제
a= [1,2,3]
a[1:3]= []  # 인덱스 1~2가 지워짐
print(a)

a= [1,2,3]
del a[1]  # 인덱스 1이 지워짐
print(a)
```

```python
# 리스트의 추가
a= [1,2,3]
# 리스트의 마지막 위치
a.append(4)
print(a)

# 리스트의 마지막 위치에 하위리스트 추가
a.append([5,6])
print(a)
```

```python
# 리스트의 정렬
a= [1,4,3,2]

a.sort()   # 리스트 정렬함수 list()
print(a)

a.reverse()
print(a)

a= ['a','b','c']
a.reverse()
print(a)
```

```python
a=[1,2,3]
print(a.index(3))  # 리스트 a에서 3이 위치한 인덱스값
print(a.index(1))  # 리스트 a에서 1이 위치한 인덱스값 -> 존재하지 않은 값인 경우 에러
print(a.index(0))
```

```python
# 리스트에 요소 삽입(insert)
a= [1,2,3]
print(a)
# 인덱스 0에 4 insert => insert(인덱스번호, 인덱스에 들어갈 값)
a.insert(0,4)
print(a)
```

```python
# 리스트 요소 제거 (remove)
a= [1,2,3,1,2,3]
# 첫번째 값만 제거
a.remove(3)
print(a)
```

```python
# 리스트 요소 끄집어내기 (pop)
# pop(): 리스트의 맨 마지막 요소를 끄집어내어 돌려줌

a=[1,2,3]
b=a.pop()
print(a)
print(b)
```

```python
# 리스트에 포함된 요소 x의 개수 세기 (count)
a=[1,2,3,1]
b= a.count(1)  # 1이란 값이 몇개?
print(b)
```

```python
# 리스트 확장 (extend)
# extend(x)에서 x에는 리스트만 올 수 있으며 원래의 a 리스트를 확장함
a= [1,2,3]
a.extend([4,5])
print(a)
```

```python
# in과 not in
a= [1,2,3,4,5]

# a에 2가 포함되었는가?
result= 2 in a
print(result)

# a에 2가 포함되어 있지 않은가?
result= 2 not in a
print(result)
```

### 튜플 자료형
### 리스트는 대괄호[]로 감싸고 튜플은 소괄호 ()로 감싼다.
### 리스트는 수정,삭제가 되지만 튜플은 수정, 삭제가 불가능함

### t1 = (1, 2, 'a', 'b')

### t1[0] = 'c'

```python
# 튜플은 ()로 감싼다.
t1=(1,2,'a','b')
print(t1[0]) 

# 튜플은 내용 변경이 안됨
# t1[0]= 10   
```

```python
t1= (1,2,'a','b')
print(type(t1))
print(t1)  # 튜플의 모든 요소 출력 
print(t1[0])  # 튜플의 첫번째 요소

print(t1[1:])  # 인덱스 1부터 끝까지

t2= (3,4)
print(t1+t2)  # 튜플 합치기
print(t1)
print(t2)
print(t2*3)  # 튜플 곱하기(반복)
```

```python
# 튜플 풀기
t1= (1,2,3)
# 튜플의 값을 각각의 변수에 저장
n1,n2,n3 = t1
print("{0}{1}{2}".format(n1,n2,n3))

# 변수값 교환
n1,n2 = n2,n1
print('{}{}'.format(n1,n2))
```

```python
# 튜플과 리스트의 변환
my_list = [1,2,3]
my_tuple = ('x','y','z')
# 리스트를 튜플로 복사
print(tuple(my_list))
# 튜플을 리스트로 복사
print(list(my_tuple))
print(my_list)
print(my_tuple)
```

### 딕셔너리 자료형
### key와 value를 한쌍으로 갖는 자료형 (자바의 map과 비슷함)
### {key:value}

```python
a={1:'a'}
a[2]= 'b'
print(a)
print(type(a))
```

```python
a= {'name':'김철수','phone':'01099125482','birth':'901231'}
print(a)
print(a['name'])  # 딕셔너리['key']
print(a.get('name'))  # 딕셔너리.get('key')
print(a['phone'])
print(a.get('phone'))
print(a['birth'])
print(a.get('birth'))
```

```python
# key 리스트 만들기(keys)
a= {'name':'김철수','phone':'01099125482','birth':'901231'}
print(a.keys())  # 딕셔너리에 key값만 출력

# for 개별변수 in 집합변수
for k in a.keys():
    print(k)

# keys 객체를 리스트 또는 튜플로 변환
b= list(a.keys())
print(b)

c= tuple(a.keys())
print(c)
```

```python
# value 리스트 만들기 (values)
a= {'name':'김철수','phone':'01099125482','birth':'901231'}
b= a.values()
print(b)
print(type(b))
for val in b:
    print(val)
```

```python
# 딕셔너리의 아이템
a= {'name':'김철수','phone':'01099125482','birth':'901231'}
b= 'name' in a  # a 딕셔너리에 key가 name인 요소가 있는가
print(b)
b= 'email' not in a  # a 딕셔너리에 key가 email인 요소가 없는가
print(b)
```

### 집합 자료형
#### 중복값을 허용하지 않음, 순서가 없음

```python
s1= set([1,2,3])
print(s1)
```

```python
# 교집합
s1= set([1,2,3,4,5,6])
s2= set([4,5,6,7,8,9])
s3= s1 & s2
print(s3)
s3= s1.intersection(s2)  # intersection() 함수 표현가능
print(s3)
```

```python
# 합집합
s4= s1 | s2
print(s4)
s4= s1.union(s2)   # A.union(B) 함수로 표현가능
print(s4)
```

```python
# 차집합
s5= s1-s2
print(s5) 
s5= s1.difference(s2)  # A.differene(B) 함수로 표현가능
print(s5)
```

```python
# 값 1개 추가하기 (add)
s1= set([1,2,3])
s1.add(4)
print(s1)

# 값을 여러개 추가하기 (update)
s1= set([1,2,3])
s1.update([4,5,6])
print(s1)
```

```python
# 특정값 제거하기 (remove)
s1= set([1,2,3])
s1.remove(2)
print(s1)
```

### 날짜

```python
from datetime import date,time,datetime,timedelta

# today() : 오늘 날짜를 구하는 함수
today= date.today()
print('today: {0}'.format(today))

print('연도: {0}'.format(today.year))
print('월: {0}'.format(today.month))
print('일: {0}'.format(today.day))

# 날짜 시간 구하기
timenow= datetime.today()
print(timenow)
print('날짜시간: {0}'.format(timenow))
```

```python
today= date.today()
# 하루전
days= timedelta(days=-1)
print('어제: {0}'.format(today+days))

# 일주일 전
days= timedelta(days=-7)
print('1주전: {0}'.format(today+days))

# 1개월 전
days= timedelta(days=-30)
print('1개월전: {0}'.format(today+days))

# 1개월 후
days= timedelta(days=30)
print('1개월후: {0}'.format(today+days))

# 8시간 전
hours= timedelta(hours=-8)
print('{0!s}일 {1!s}초'.format(hours.days, hours.seconds))
```

```python
# 날짜 출력 형식
today= datetime.today()

# Y: 4자리 연도, y: 2자리 연도, H: 24시간제, I: 12시간제
print('{}'.format(today.strftime('%m/%d/%Y')))
print('{}'.format(today.strftime('%Y-%m-%d %H:%M:%S')))
print('{}'.format(today.strftime('%y-%m-%d %I:%M:%S')))
```
